```python
import requests
import pandas as pd
import time

# Step 1: Get the genre map
genre_url = "https://api.themoviedb.org/3/genre/movie/list?api_key=ee78a13f6557e14dacc746275efa79f3&language=en-US"
genre_response = requests.get(genre_url)
genre_data = genre_response.json()["genres"]

# Build a dictionary: {id: name}
genre_map = {g["id"]: g["name"] for g in genre_data}

# Step 2: Collect movie data
all_dfs = []

for i in range(1, 525):
    try:
        url = f'https://api.themoviedb.org/3/movie/top_rated?api_key=ee78a13f6557e14dacc746275efa79f3&language=en-US&page={i}'
        response = requests.get(url)
        response.raise_for_status()
        data = response.json().get("results", [])
        if data:
            temp_df = pd.DataFrame(data)[["title", "overview", "genre_ids"]]
            all_dfs.append(temp_df)
        time.sleep(0.25)
    except Exception as e:
        print(f"Error on page {i}: {e}")

df = pd.concat(all_dfs, ignore_index=True)

# Step 3: Replace genre IDs with names
def map_genre_ids(ids):
    return [genre_map.get(i, "Unknown") for i in ids]

df["genres"] = df["genre_ids"].apply(map_genre_ids)
df.drop(columns=["genre_ids"], inplace=True)

# Step 4: (Optional) Expand into a comma-separated string
df["genres"] = df["genres"].apply(lambda x: ", ".join(x))

# ✅ Final dataframe
print(df.head())
```

Remove stop words

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['overview'].apply(remove_stopwords)
```

Remove emoji

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Loved the movie. It was 😘😘")
```

Stemmer

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
