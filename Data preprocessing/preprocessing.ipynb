```python
import requests
import pandas as pd
import time

# Step 1: Get the genre map
genre_url = "https://api.themoviedb.org/3/genre/movie/list?api_key=ee78a13f6557e14dacc746275efa79f3&language=en-US"
genre_response = requests.get(genre_url)
genre_data = genre_response.json()["genres"]

# Build a dictionary: {id: name}
genre_map = {g["id"]: g["name"] for g in genre_data}

# Step 2: Collect movie data
all_dfs = []

for i in range(1, 525):
    try:
        url = f'https://api.themoviedb.org/3/movie/top_rated?api_key=ee78a13f6557e14dacc746275efa79f3&language=en-US&page={i}'
        response = requests.get(url)
        response.raise_for_status()
        data = response.json().get("results", [])
        if data:
            temp_df = pd.DataFrame(data)[["id", "overview", "genre_ids"]]
            all_dfs.append(temp_df)
        time.sleep(0.25)
    except Exception as e:
        print(f"Error on page {i}: {e}")

df = pd.concat(all_dfs, ignore_index=True)

# Step 3: Replace genre IDs with names
def map_genre_ids(ids):
    return [genre_map.get(i, "Unknown") for i in ids]

df["genres"] = df["genre_ids"].apply(map_genre_ids)
df.drop(columns=["genre_ids"], inplace=True)

# Step 4: (Optional) Expand into a comma-separated string
df["genres"] = df["genres"].apply(lambda x: ", ".join(x))

# ✅ Final dataframe
print(df.head())
```
